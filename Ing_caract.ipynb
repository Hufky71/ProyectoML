```python
import pandas as pd

def detect_outliers(df):
    # Crear un DataFrame vacío para almacenar los outliers
    outliers = pd.DataFrame(columns = df.columns)

    # Iterar sobre cada columna
    for col in df.columns:
        if df[col].dtype.kind in 'bifc':  # Solo columnas numéricas
            # Calcular el IQR
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1

            # Definir los límites para los outliers
            lower_limit = Q1 - 1.5 * IQR
            upper_limit = Q3 + 1.5 * IQR

            # Identificar los outliers
            results = df[(df[col] < lower_limit) | (df[col] > upper_limit)].copy()
            results['Outlier'] = col
            outliers = pd.concat([outliers, results], ignore_index=True)

    return outliers

# Uso de la función
outliers = detect_outliers(df)
```
